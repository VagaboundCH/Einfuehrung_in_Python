## 4.3 Datentypen
### 4.3.1 Ganze Zahlen
Im Gegensatz zu anderen Programmiersprachen können z.B. Integers unbegrenzt gross sein. Sie besitzen also keinen Maximalwert.
Binär, Oktal und Hex können wie folgt generiert werden: 

```python
print(0b10, bin(2))
print(0o10, oct(8))
print(0x10, hex(16))
```

### 4.3.2 Fliesskommazahlen
Alle Fliesskommazahlen sind in Python floats.  
$3.14*10^{2}$ und $5^{2}$ kann wie folgt generiert werden:

```python
print(3.14e2, '\n', 5**3)
```

### 4.3.5 Komplexe Zahlen
Mit Komplexe Zahlen kann wie folgt gerechnet werden:

```python
y = 2 - 5j
z = 4 + 2j
print(y+z)
```

# 6 Dictionaries
## 6.2 Definition und Benutzung
Dictionaries werden mit geschweiften Klammern gebildet und bestehen aus einem Schlüsselwertpaar.
Der Schlüssel und Wert wird jeweils mit einem Doppelpunkt getrennt.

```python
en_de = {"red" : "rot", "green" : "grün", "blue" : "blau", "yellow" : "gelb"}
print(en_de)
```

```python
en_de["red"]
```

Mithilfe eines deutsch-französischen Dictionary lässt sich nun auch von englisch auf französisch übersetzen.

```python
de_fr = {"rot" : "rouge", "grün" : "vert", "blau" : "blue", "gelb" : "jaune"}
de_fr[en_de["red"]]
```

## 6.3 Fehlerfreie Zugriffe auf Dictionaries
Mit dem Schlüsselwort *in* kann geprüft werden, ob ein Index in einem Dictionary vorkommt:

```python
"red" in en_de
```

Mithilfe der Funktion *get()* kann auch auf ein Element zugegriffen werden.

```python
en_de.get("red")
```

Dictionaries, deren Schlüssel nicht aus Strings besthen, lassen sich wie folgt definieren:

```python
en_de_1 = dict(red = "rot", green = "grün", blue = "blau", yellow = "gelb")
en_de_1
```

## 6.4 Zulässige Typen für Schlüssel und Werte
Es lassen sich auch andere Typen als Schlüssel verwenden, diese müssen jedoch immer unveränderlich sein, sonst erhält man eine Fehlermeldung.

```python
staedte_GPS = {(52.52, 13.404) : "Berlin"}
print(staedte_GPS)
staedte_GPS_1 = {[52.52, 13.404] : "Berlin"}
```

Tupel als Schlüsse sind in ordnung, Listen jedoch nicht.

## 6.5 Verschachtelte Dictionaries (Nested Dictionaries)
Wir definieren ein Dictonary von Dictionaries:

```python
dictionaries = {"en_de" : en_de, "de_fr" : de_fr}
print(dictionaries["de_fr"]["blau"])
```
